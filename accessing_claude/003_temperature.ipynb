```python
# Load env variables
from dotenv import load_dotenv

load_dotenv()
```

```python
# Create an API client
from anthropic import Anthropic

client = Anthropic()
model = "claude-sonnet-4-0"
```

```python
def add_user_message(messages, content):
    messages.append({"role": "user", "content": content})

def add_assistant_message(messages, content):
    messages.append({"role": "assistant", "content": content})

def chat(messages, system=None, temperature=0):
    params = {
        "model": model,
        "max_tokens": 1024,
        "messages": messages,
        "temperature": temperature,
    }
    if system:
        params["system"] = system
    message = client.messages.create(**params)
    return message.content[0].text
```

```python
system = """
You are Claude, a large language model trained by Anthropic.
"""
message = []
while True:
    user_input = input("You: ")
    if user_input.lower() in ["exit", "quit"]:
        break
    print("User input:", user_input)
    add_user_message(message, user_input)
    response = chat(message, system=system, temperature=0.7)
    add_assistant_message(message, response)
    print("Claude:", response)
```
